```python
import openpyxl
from openpyxl.drawing.image import Image as XLImage
from PIL import Image
import os
import io  # 임시파일 대신 메모리 버퍼 사용
import re

# 경로 설정
excel_path = "C:/Users/jusun/OneDrive/문서/카카오톡 받은 파일/웰스토리 20250717.xlsx"
output_path = "웰스토리_이미지삽입완료.xlsx"
이미지폴더 = "C:/Users/jusun/OneDrive/문서/카카오톡 받은 파일/해오라기"

# 이미지파일 전체 리스트 가져오기
image_list = os.listdir(이미지폴더)

# 엑셀 열기
wb = openpyxl.load_workbook(excel_path)
ws = wb.active

# E열 기준으로 이미지 삽입
for row in range(2, ws.max_row + 1):
    상품명 = ws[f'E{row}'].value

    if 상품명:
        target_img = None
        for img_file in image_list:
            if 상품명 in img_file:
                target_img = img_file
                break

        if target_img:
            img_full_path = os.path.join(이미지폴더, target_img)
            try:
                img = Image.open(img_full_path)
                img = img.resize((300, 300))

                # 메모리 버퍼에 PNG 형태로 저장
                img_bytes = io.BytesIO()
                img.save(img_bytes, format='PNG')
                img_bytes.seek(0)

                # openpyxl Image로 변환
                xl_img = XLImage(img_bytes)
                ws.add_image(xl_img, f'G{row}')

                print(f"{row}행 이미지 '{target_img}' 삽입 완료")
            except Exception as e:
                print(f"{row}행 이미지 오류: {e}")
        else:
            print(f"{row}행 '{상품명}'에 해당하는 이미지 없음")

# 결과 저장
wb.save(output_path)
print("엑셀 이미지 삽입 완료:", output_path)
```
